```python
# 1. Use matplotlib to plot the following equation: 'y = x^2 - x + 2'
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.
```

```python
import matplotlib.pyplot as plt
import math as m
```

```python
plt.figure(figsize = (8,5))

x = range(-50, 51)
y = [n ** 2 - n + 2 for n in x]

plt.plot(x,y)

plt.grid( True, ls = ':')

plt.annotate('origin', xy=(0,0), xytext=(0,250), arrowprops={'facecolor': 'blue'})

plt.show()
```

```python
# 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
# -a. 'y = \sqrt{x}'
# -b. 'y = x^3'
# -c. 'y = 2^x'
# -d. 'y = 1/(x+1)'
# You can use functions from the math module to help implement some of the equations above.
```

```python
# 2a. 'y = \sqrt{x}'

plt.figure(figsize = (8,5))

x = range(0, 51)
y = [m.sqrt(n) for n in x]

plt.plot(x,y)

plt.title('$y=\sqrt{x}$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$y=\sqrt{x}$', fontsize = 14)
plt.grid(True, ls = ':')

plt.show()
```

```python
# 2b. 'y = x^3'

plt.figure(figsize = (8,5))

x = range(0, 51)
y = [n ** 3 for n in x]

plt.plot(x,y)

plt.title('$y = x^3$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^3$', fontsize = 14)
plt.grid(True, ls = ':')

plt.show()
```

```python
# 2c. 'y = 2^x'

plt.figure(figsize = (8,5))

x = range(0, 51)
y = [2 ** n for n in x]

plt.plot(x,y)

plt.title('$y = 2^x$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$2^x$', fontsize = 14)
plt.grid(True, ls = ':')

plt.show()
```

```python
# 2d. 'y = 1 / (x + 1)'

plt.figure(figsize = (8,5))

x = range(0, 51)
y = [1/(n+1) for n in x]

plt.plot(x,y)

plt.title('$y = 1 / (x + 1)$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$1/(x+1)$', fontsize = 14)
plt.grid(True, ls = ':')

plt.show()
```

```python
# 3. Combine the figures you created in the last step into one large figure with 4 subplots.
```

```python
x = [x * 0.25 for x in range(0,50)]

y1 = [m.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]
y4 = [1 / (n + 1) for n in x]

plt.figure(figsize = (10,8))

plt.suptitle('One Large Figure with 4 Subplots', fontsize = 16)

plt.subplot(2,2,1)
plt.plot(x,y1, marker = '*', color = 'red', ls = '-')
plt.title('$y=\sqrt{x}$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$y=\sqrt{x}$', fontsize = 14)
plt.grid(True, ls = ':')

plt.subplot(2,2,2)
plt.plot(x,y2, marker = 's', color = 'orange', ls = '')
plt.title('$y = x^3$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$x^3$', fontsize = 14)
plt.grid(True, ls = ':')

plt.subplot(2,2,3)
plt.plot(x,y3, marker = 'x', color = 'green', ls = ':')
plt.title('$y = 2^x$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$2^x$', fontsize = 14)
plt.grid(True, ls = ':')

plt.subplot(2,2,4)
plt.plot(x,y4, marker = 'o', color = 'blue')
plt.title('$y = 1/(x+1)$', fontsize = 14)
plt.xlabel('$x$', fontsize = 14)
plt.ylabel('$1/(x+1)$', fontsize = 14)
plt.grid(True, ls = ':')

plt.tight_layout()

plt.show()
```

```python
# 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a 
# different color for the points. Be sure to include a legend and an appropriate title for the figure.
```

```python
x = [x for x in list(range(1,10))]

y1 = [m.sqrt(n) for n in x]
y2 = [n ** 3 for n in x]
y3 = [2 ** n for n in x]
y4 = [1 / (n + 1) for n in x]

plt.figure(figsize = (16,9))

plt.plot(x,y1, color = 'red', label = '$y=\sqrt{x}$')
plt.plot(x,y2, color = 'orange', label = '$y = x^3$')
plt.plot(x,y3, color = 'green', label = '$y = 2^x$')
plt.plot(x,y4, color = 'blue', label = '$y = 1/(x+1)$')

plt.title('4 Equations in 1 Figure', fontsize = 14)
plt.grid(True, ls = ':')
plt.legend()

plt.show()
```
